## 2. Euclidean distance (*)

The formula for Euclidean distance in 2D between $P: (p_1, p_2)$ and $Q: (q_1, q_2)$  is: 

$d(P,Q) = \sqrt{(p_1-q_1)^2 + (p_2-q_2)^2}$

&nbsp; a) &nbsp; Create a function that takes two points as input parameters and return the Euclidean between them. (*)

&nbsp; b) &nbsp; Let the user input two points. Call the function using the users input points. (*)

&nbsp; c) &nbsp; Use your function to calculate distances between the origin (0, 0) and each of these points: (10, 3), (-1, -9), (10, -10), (4, -2), (9, -10). (*)

<details>

<summary>Hint</summary>

Use a **for** statement

</details>
<br>
<details>

<summary>Answer</summary>

c)

```
[10.4, 9.1, 14.1, 4.5, 13.5]
```

</details>


````python
#Uppgift A
import math

def distance(p1,p2,q1,q2):
    euclidean = math.sqrt((p1-q1)**2 + (p2-q2)**2)
    return euclidean

print(distance(-4,2,5,1))
````

````python
#Uppgift A
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,400)
f =lambda x: (x**2-3)
y =f(x)

plt.plot(x,y, label="f(x) = x^2 - 3")
plt.xlabel("x")
plt.ylabel("y")
plt.title("f(x) = x^2 - 3")
plt.legend()
plt.show()


````

````python
#Uppgift B
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,400)
g =lambda x: (4*x-7)
y =g(x)

plt.plot(x, y, color="red", label="g(x) = 4x-7")
plt.xlabel("x")
plt.ylabel("y")
plt.title("g(x) = 4x-7")
plt.legend()
plt.show()


````

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,400)
g =lambda x: (4*x-7)
f =lambda x: (x**2-3)
y1 =g(x)
y2 =f(x)

plt.plot(x, y1, color="red", label="g(x) = 4x-7")
plt.plot(x, y2, color="blue", label="f(x) = x^2-3")
plt.xlabel("x")
plt.ylabel("y")
plt.axvline(0, color="black", linewidth=1)
plt.axhline(0, color="black", linewidth=1)
plt.title("g(x), f(x)")
plt.legend()
plt.grid(True)
plt.show()
````

## 4. Name cleaner (*)

Create a function that takes a name as an input and:

- removes all leading and trailing blank spaces
- make capitalize the first character of each name, and make the rest lowercase

Use your function on this list of strings: 

```
["  MaRcUs ", " iDA aNderSon", "OLOF Olofsson            "  ]
```


<details>

<summary>Hint</summary>
Use 

- [title() method][str_title]
- [strip() method][str_strip]
- [join() method][str_join]
- [split() method][str_split]

[str_title]: https://www.w3schools.com/python/ref_string_title.asp

</details>
<br>
<details>

<summary>Answer</summary>

```
Marcus
Ida Anderson
Olof Olofsson
```

</details>


````python
names= ["  MaRcUs ", " iDA aNderSon", "OLOF Olofsson            "  ]
result = []

def name_cleaner(text):
    text = text.strip()
    text = text.capitalize()
    return text

for name in names:
    result.append(name_cleaner(name))

print(result)
    
````
